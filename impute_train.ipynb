```python
import pandas as pd
import numpy as np
from ycimpute.imputer import iterforest
```

```python
file = 'train_clean.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.drop(labels=['Id', 'Unnamed: 0', 'FireplaceQu'], axis=1, inplace=True)
```

```python
df_copy = df.copy()
```

```python
no_missingdata = df_copy.columns[(df_copy.isnull().any() != True)].tolist()
no_missingdata
```

```python
missingdata = df_copy.columns[df_copy.isnull().any()].tolist()
missingdata
```

```python
len(missingdata)
```

## Impute `LotFrontage`

```python
column = 'LotFrontage'
```

```python
data_ = no_missingdata
data_
```

```python
data_.append(column)
```

```python
data_
```

```python
index_na = pd.isnull(df_copy[data_]).any(1).nonzero()[0].tolist()
index_na
```

```python
(len(index_na)/df_copy.shape[0])*100
```

```python
df_data_ = df_copy[data_]
df_data_.head()
```

```python
df_data_[column].unique()
```

```python
impute_ = iterforest.IterImput().complete(df_data_.as_matrix())
```

```python
impute_
```

```python
df_impute_ = pd.DataFrame(impute_)
df_impute_.head()
```

```python
df_impute_.shape
```

```python
df_impute_.isnull().any().any()
```

```python
df_impute_ = round(df_impute_)
```

```python
df_impute_.loc[index_na]
```

```python
df_copy[column] = df_impute_[61]
```

```python
df_copy[column].unique()
```
